# Data Analysis & Visualization


```python
canceled_perc = df['is_canceled'].value_counts(normalize = True)
print(canceled_perc)

plt.figure(figsize =(5,6))
plt.title('Reservation Status')
colors = ['lightgreen', 'red']
plt.bar(['Not Canceled', 'Canceled'],df['is_canceled'].value_counts(), edgecolor='k', width=0.7, color=colors)
plt.show()
```

```python
plt.figure(figsize=(9, 7))
ax = sns.countplot(x = 'hotel', hue ='is_canceled', data = df, palette ='BuGn')
legend_labels, _ = ax.get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status by Hotel Type', size=25)
plt.xlabel('Hotel', fontweight='bold')
plt.ylabel('Number of Reservations')
legend_labels, _ = ax.get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status by Hotel Type', size=25)
plt.xlabel('Hotel', fontweight='bold')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel'] 
resort_hotel ['is_canceled'].value_counts(normalize =True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel'] 
city_hotel  ['is_canceled'].value_counts(normalize =True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
plt.figure(figsize = (25,8))
plt.title('Avg Daily Rate in City & Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = pd.to_datetime(df['reservation_status_date']).dt.month
custom_palette = {'darkblue'}
plt.figure(figsize=(16, 8))
ax = sns.countplot(x='month', hue='is_canceled', data=df,)
plt.title('Months with the Highest Cancellation Rates' ,fontsize = 25)
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per month' ,fontsize = 25)
ax = sns.barplot(x='month', y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())                                        
plt.show()
```

```python
canceled_data = df[df['is_canceled'] == 1]
top_10_country = canceled_data ['country'].value_counts()[:10]
plt.figure(figsize=(8, 8))
plt.title('Top 10 Countries with Reservations Canceled' ,fontsize = 25)
plt.pie(top_10_country , autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace =True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data =  df[df['is_canceled'] == 0]
not_canceled_adr  = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_adr.reset_index(inplace = True)
not_canceled_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize=(25, 8))
plt.title('Avg Daily Rate')
plt.plot(not_canceled_adr['reservation_status_date'], not_canceled_adr['adr'], label='not_canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='Canceled')
plt.legend(fontsize=20)
plt.show() 
```

```python
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] > '2016') & (canceled_df_adr['reservation_status_date'] < '2017-09')]
not_canceled_adr = not_canceled_adr[(not_canceled_adr['reservation_status_date'] > '2016') & (not_canceled_adr['reservation_status_date'] < '2017-09')]
```

```python
plt.figure(figsize=(20, 6))
plt.title("Average Daily Rate", fontsize=30)
plt.plot(not_canceled_adr['reservation_status_date'], not_canceled_adr['adr'], label='not cancelled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='cancelled')
plt.legend(fontsize=20)
plt.show()
```
